```python
def round_to_nearest_two(number):
    """Rounds a number to the nearest multiple of 2."""
    return round(number / 2) * 2

def round_to_nearest_three(number):
    """Rounds a number to the nearest multiple of 2."""
    return round(number / 3) * 3

def round_to_nearest_four(number):
    """Rounds a number to the nearest multiple of 2."""
    return round(number / 4) * 4

def round_to_nearest_five(number):
    """Rounds a number to the nearest multiple of 2."""
    return round(number / 5) * 5

 
```

```python
tmpall['timedelta2'] = tmpall['timedelta'].apply(round_to_nearest_two)
tmpall['timedelta3'] = tmpall['timedelta'].apply(round_to_nearest_three)
tmpall['timedelta4'] = tmpall['timedelta'].apply(round_to_nearest_four)
tmpall['timedelta5'] = tmpall['timedelta'].apply(round_to_nearest_five)
```
