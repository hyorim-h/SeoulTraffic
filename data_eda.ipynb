```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
weekday = pd.read_csv('J:/아이나비_202010/hanyang_buffer_saturday.csv')
weekday
```

```python
weekday.info()
```

```python
weekday['timestamp2'] = pd.to_datetime(weekday['timestamp2'])
weekday.info()
```

```python
weekday['day_of_week'] = weekday['timestamp2'].dt.day_name()
weekday
```

```python
sns.countplot(x=weekday['day_of_week'])
```

```python
weekday['day'] = weekday['timestamp2'].dt.day
weekday
```

```python
week = weekday.drop_duplicates(['obu_id', 'group_id', 'day'])
week
```

```python
sns.countplot(x=week['day_of_week'], palette='plasma')
```

```python
sns.countplot(x=week['day'], palette='coolwarm')
```

```python
week.to_csv('J:/아이나비_202010/hanyang_buffer_holiday_dropdup.csv')
```

```python
groupby1 = week.groupby(['day_of_week', 'day'])['obu_id'].count()
groupby1
```
