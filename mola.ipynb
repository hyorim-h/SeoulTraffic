```python
link = pd.read_csv('D:/서울시신호/데이터/GIS/한양도성/한양도성진짜/lv6_link.csv', encoding='euc-kr')
link
```

```python
m_m_peak = pd.merge(left=m_peak, right=link, how='left', on='link_id')
m_m_peak
```

```python
m_m_peak.columns
```

```python
m_m_peak['max_speed']
```

```python
m_m_peak.dropna(inplace=True, subset=['up_v_link'])
m_m_peak
```

```python
m_m_peak['fft'] = (m_m_peak['length']/m_m_peak['max_speed'])*3600
m_m_peak
```

```python
m_m_peak['delta'] = pd.to_datetime(m_m_peak['out_time']) - pd.to_datetime(m_m_peak['in_time'])
```

```python
m_m_peak
```

```python
def sec(data):
    sec = data.seconds
    return sec
```

```python
m_m_peak['delta2'] = m_m_peak['delta'].apply(sec)
m_m_peak
```

```python
m_m_peak['cong'] = m_m_peak['delta2']/m_m_peak['fft']
m_m_peak
```
