```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# 예시 데이터 생성 (중복된 x 값을 포함)
# y = df.obu_id
# x = df.our_speed

# 데이터프레임 생성
df = pd.DataFrame({'x': x, 'y': y})

# 중복된 x 값에 대해 y 값의 평균 계산
df_averaged = df.groupby('x')['y'].mean().reset_index()

# 보간 스플라인 생성
spline = make_interp_spline(df_averaged['x'], df_averaged['y'], k=3)

# 보간 스플라인을 계산할 x 값 생성
x_smooth = np.linspace(df_averaged['x'].min(), df_averaged['x'].max(), 100)

# 산점도와 감싸는 곡선 그리기
plt.scatter(df_averaged['x'], df_averaged['y'], label='Data Points')
plt.plot(x_smooth, spline(x_smooth), label='Smoothed Curve', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
