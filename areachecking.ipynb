```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플 데이터
labels = ['A', 'B', 'C', 'D', 'E']
data1 = [3, 4, 6, 2, 5]
data2 = [200, 150, 50, 300, 100]

fig, ax1 = plt.subplots()

# 첫 번째 데이터에 대한 막대 그래프
bars1 = ax1.bar(labels, data1, color='b', label='Data 1', alpha=0.6)

# 두 번째 y 축 생성
ax2 = ax1.twinx()

# 두 번째 데이터에 대한 막대 그래프
bars2 = ax2.bar(labels, data2, color='r', label='Data 2', alpha=0.3)

# 축과 레이블 설정
ax1.set_xlabel('Label')
ax1.set_ylabel('Data 1', color='b')
ax2.set_ylabel('Data 2', color='r')

# 레전드 추가
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플 데이터
data1 = np.random.randn(1000) + 2  # 평균이 2인 데이터
data2 = np.random.randn(1000) * 5  # 표준 편차가 5인 데이터

fig, ax1 = plt.subplots()

# 첫 번째 데이터에 대한 히스토그램
color1 = 'blue'
n1, bins1, patches1 = ax1.hist(data1, bins=50, color=color1, alpha=0.6, label='Data 1')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count of Data 1', color=color1)
ax1.tick_params(axis='y', labelcolor=color1)

# 두 번째 y 축 생성
ax2 = ax1.twinx()

# 두 번째 데이터에 대한 히스토그램
color2 = 'red'
n2, bins2, patches2 = ax2.hist(data2, bins=50, color=color2, alpha=0.3, label='Data 2', rwidth=0.8)
ax2.set_ylabel('Count of Data 2', color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

# 레전드 추가
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 샘플 데이터 생성
data = {
    'a': np.random.randn(1000),
    'b': np.random.randn(1000) + 2
}
df = pd.DataFrame(data)

# 히스토그램 그리기
plt.figure(figsize=(10, 5))
plt.hist(df['a'], bins=30, alpha=0.5, label='a')
plt.hist(df['b'], bins=30, alpha=0.5, label='b')
plt.title('Histogram of a and b')
plt.legend()
plt.show()

# KDE 그리기
plt.figure(figsize=(10, 5))
sns.kdeplot(df['a'], shade=True, label='a')
sns.kdeplot(df['b'], shade=True, label='b')
plt.title('KDE of a and b')
plt.legend()
plt.show()
```

```python
# b의 평균을 계산 (이 예에서는 a의 값이 모두 고유하므로 원래 b의 값이 그대로 사용됩니다)
avg_b = df.groupby('a')['b'].mean()

avg_b.plot(kind='bar')
plt.xlabel('a')
plt.ylabel('Average of b')
plt.title('Bar Plot of a vs. Average of b')
plt.grid(True, axis='y')
plt.show()
```
