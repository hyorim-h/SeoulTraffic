```python
import pandas as pd
pd.options.display.float_format = '{: .7f}'.format
```

```python
df = pd.read_csv('J:/hyorim/od_area_point.csv')
df
```

```python
std_lon = df['org_longitude'][0]
std_lon
```

```python
std_lat = df['org_latitude'][0]
std_lat
```

```python
lon = []
lat = []
```

```python
for i in range(0,100):
    lon.append(round(std_lon + i * 0.0000001, 8))

lon
```

```python
for i in range(0,100):
    lat.append(round(std_lat + i * 0.0000001, 8))

lon
```

```python
for i in range(100):
    for j in range(100):
        new_data = {
            'od_id': '108420220926215910847S00',
            'org_zone_no' : '11110_11680',
            'org_longitude' : lon[i],
            'org_latitude' : lat[j],
            'org_datetime' : '2.022090e+13',
            'dst_longitude' : 127.03889,
            'dst_latitude' : 37.39487,
            'dist_datetime' : '2.022090e+13',
            'dist' : 12977
        }
        df = df.append(new_data, ignore_index=True)
        print(lon[i], lat[j])

df
```
